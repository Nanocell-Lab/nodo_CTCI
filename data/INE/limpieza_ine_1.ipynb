### Diccionario Columnas

```python
descriptor = pd.read_excel('descriptor-de-variables-sector-empresas.xlsx', sheet_name='Descriptor_Empresas')
descriptor['identificador'] = (descriptor['MÓDULO'].astype(str)+descriptor['VARIABLE'].astype(str))\
                            .apply(lambda x: x.replace('-',""))

descriptor = descriptor.drop(columns=['MÓDULO', 'VARIABLE'])
descriptor['identificador'] = descriptor['identificador'].str.lower()


transpuesta = descriptor.transpose().copy()
transpuesta.columns = transpuesta.iloc[2]
transpuesta.drop(transpuesta.index[1])
transpuesta = transpuesta.drop(transpuesta.index[1])
transpuesta = transpuesta.drop(transpuesta.index[1])


diccionario_columns = transpuesta.to_dict('r')[0]

columns_empresa = diccionario_columns

columns_empresa
```
